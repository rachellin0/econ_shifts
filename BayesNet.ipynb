```python
import numpy as np
import pandas as pd
from sklearn import svm, metrics
import matplotlib.pyplot as plt
#Load Data
msa_data = pd.read_csv('/content/drive/MyDrive/MLC Final Project/4_MSA_LQ.csv')
pop_data = pd.read_csv('/content/drive/MyDrive/MLC Final Project/1_PopChange_10_16.csv')
```

```python
msa_data.describe()
```

```python
for i in msa_data.columns:
  if i != 'MSA':
    for j in range(len(msa_data[i])):
      if msa_data[i][j] >= 1:
        msa_data[i][j] = 1
      else:
        msa_data[i][j] = 0
```

```python
pop_data['PopCategory'] = -100
```

```python
for i in range(len(pop_data)):
  if ((pop_data['PopChange'][i] <= 0.5) & (pop_data['PopChange'][i] > -0.5)):
    pop_data['PopCategory'][i] = 0
  elif ((pop_data['PopChange'][i] > 0.5) & (pop_data['PopChange'][i] <= 1)):
    pop_data['PopCategory'][i] = 1
  elif (pop_data['PopChange'][i] > 1):
    pop_data['PopCategory'][i] = 2
  elif ((pop_data['PopChange'][i] < -0.5) & (pop_data['PopChange'][i] >= -1)):
    pop_data['PopCategory'][i] = -1
  elif (pop_data['PopChange'][i] < -1):
    pop_data['PopCategory'][i] = -2
```

```python
#Merge MSA data and Population data
df_std = msa_data.merge(pop_data[['msaID','PopChange', 'PopCategory']], left_on='MSA',right_on='msaID').drop('msaID',axis=1)
df_std
```

```python
from sklearn.model_selection import train_test_split
df_std_train, df_std_test = train_test_split(df_std,random_state=999,test_size=0.3)
```

```python
#PC
from pgmpy.estimators import PC
from pgmpy.estimators.CITests import chi_square

est = PC(df_std_train)

skel, separating_sets = est.build_skeleton(significance_level=0.01)
print("Undirected edges: ", skel.edges())

pdag = est.skeleton_to_pdag(skel, separating_sets)
print("PDAG edges:       ", pdag.edges())

model = est.estimate()
print("DAG edges:        ", model.edges())
```

```python
from pgmpy.models import BayesianModel
model1 = BayesianModel([('Management', 'Retail_Trade'), ('Information', 'Professional_Services'), ('Real_Estate', 'PopCategory'), ('Real_Estate', 'Construction'), ('Construction', 'PopCategory'), ('Educational', 'Agriculture'), ('Manufacturing', 'Real_Estate')])  # fruit -> tasty <- siz
```

```python
#This is a somewhat arbitrarily 'trimmed' version of the above net that I found worked

# model = BayesianModel([('Management', 'Retail_Trade'), ('Real_Estate', 'new_clf'), ('Manufacturing', 'Real_Estate'), ('Health_Care', 'new_clf')])
```

```python
from pgmpy.estimators import BayesianEstimator
model1.fit(df_std_train, estimator=BayesianEstimator, prior_type='K2')
for cpd in model1.get_cpds():
    print("CPD of {variable}:".format(variable=cpd.variable))
    print(cpd)
```

```python
print(df_std.head())
df_std.columns
# model.predict(df_std.loc[:4,['Utilities','Management']])
model1.predict(df_std_test.loc[:,['Information','Professional_Services', 'Management', 'Retail_Trade', 'Real_Estate', 'Construction', 'Educational', 'Agriculture']])

```

```python
predicted_train = model1.predict(df_std_train.loc[:,['Information','Professional_Services', 'Management', 'Retail_Trade', 'Real_Estate', 'Construction', 'Educational', 'Agriculture']])
print("In sample:",(df_std_train.loc[:,'PopCategory'].reset_index(drop=True)==predicted_train['PopCategory']).mean())


predicted_test = model1.predict(df_std_test.loc[:,['Information','Professional_Services', 'Management', 'Retail_Trade', 'Real_Estate', 'Construction', 'Educational', 'Agriculture']])
print("Out of sample:",(df_std_test.loc[:,'PopCategory'].reset_index(drop=True)==predicted_test['PopCategory']).mean())
```
